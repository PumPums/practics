```
jsigm = 0.2;
jb = 0.06;
jm = 0.003;
```

```
attractor[ix1_, ix2_, ix3_, isigm_, im_, ib_, ir_, tk_]:= NDSolve[{x1'[t] == isigm*(x2[t]- x1[t]),
                                    x2'[t] == ir*x1[t] - im*x2[t] - x1[t]*x3[t],
                                    x3'[t] == -ib*x3[t] + x1[t]*x2[t],
                                    x1[0] == ix1, 
                                    x2[0] == ix2, 
                                    x3[0] == ix3},
                                    {x1[t], x2[t], x3[t]},
                                    {t, 0, tk}, Method -> "StiffnessSwitching"];
```

```
func[tk_] := attractor[0.005, 0.005, 0.3, jsigm, jb, jm, rk];
```

```
FinalTime = 1000;
rk = 0.2;
sol1 = attractor[0.005, 0.005, 0.3, jsigm, jb, jm, rk, FinalTime];
sol2 = attractor[-0.005, -0.005, 0.3, jsigm, jb, jm, rk, FinalTime];
pl1 = ParametricPlot[Evaluate[{x1[t], x3[t]} /. sol1], {t, 0, FinalTime}, PlotRange -> All, AxesLabel -> {Style["x1", Medium, Bold], Style["x3", Medium, Bold]}, PlotStyle -> Blue];
pl2 = ParametricPlot[Evaluate[{x1[t], x3[t]} /. sol2], {t, 0, FinalTime}, PlotRange -> All, AxesLabel -> {Style["x1", Medium, Bold], Style["x3", Medium, Bold]}, PlotStyle -> Blue];
Show[pl1, pl2, PlotRange -> All, AxesOrigin -> {0, 0}]
```
