```python
import numpy as np
import pandas as pd
from scipy.signal import welch
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
path = 'Стабилограф/'
```

```python
# %1.Cтабилографический сигнал X                      
# 2.Cтабилографический сигнал Y                      
# 3.Баллистограмма 
```

# EO

```python
df_EO = pd.read_csv(path+'MEV_EO.txt', sep='\s+')
```

```python
df_EO.head(3)
```

```python
(Fx_EO, Px_EO) = welch(df_EO['X'], fs=50, window='hann', nperseg=1024, nfft=1024, noverlap=500)
```

```python
plt.figure(figsize=(17, 8))
plt.title('X_EO')
plt.xlabel(r'Fx_EO')
plt.ylabel(r'Px_EO')
plt.xlim([0, 5])
plt.ylim([0, 30])

plt.plot(Fx_EO, Px_EO, color='g')
```

```python
(Fy_EO, Py_EO) = welch(df_EO['Y'], fs=50, window='hann', nperseg=1024, nfft=1024, noverlap=500)
```

```python
plt.figure(figsize=(17, 8))
plt.title('Y_EO')
plt.xlabel(r'Fx_EO')
plt.ylabel(r'Px_EO')
plt.xlim([0, 5])
plt.ylim([0, 35])

plt.plot(Fy_EO, Py_EO, color='g')
```

```python
(Fb_EO, Pb_EO) = welch(df_EO['Balilstogram'], fs=50, window='hann', nperseg=1024, nfft=1024, noverlap=500)
```

```python
plt.figure(figsize=(10, 4))
plt.title('B_EO')
plt.xlabel(r'Fb_EO')
plt.ylabel(r'Pb_EO')
plt.xlim([0, 10])
plt.ylim([0, 0.0025])

plt.plot(Fb_EO, Pb_EO, color='g')
```
