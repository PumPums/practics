```python
import numpy as np
import pandas as pd
from scipy.signal import welch
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
path = 'Стабилограф/'
```

```python
# %1.Cтабилографический сигнал X                      
# 2.Cтабилографический сигнал Y                      
# 3.Баллистограмма 
```

# EO

```python
df_EO = pd.read_csv(path+'MEV_EO.txt', sep='\s+')
```

```python
df_EO.head(3)
```

```python
(Px_EO, Fx_EO) = welch(df_EO['X'], fs=50, window='hann', nperseg=1024, nfft=1024, noverlap=500)
```

```python
plt.figure(figsize=(17, 8))
plt.title('X_EO')
plt.xlabel(r'Fx_EO')
plt.ylabel(r'Px_EO')
plt.xlim([-1, 5])
plt.ylim([0, 10])

plt.scatter(Fx_EO, Px_EO, color='g', s=5)
```

```python
(Py_EO, Fy_EO) = welch(df_EO['Y'], fs=50, window='hann', nperseg=1024, nfft=1024, noverlap=500)
```

```python
plt.figure(figsize=(17, 8))
plt.title('Y_EO')
plt.xlabel(r'Fx_EO')
plt.ylabel(r'Px_EO')
plt.xlim([-1, 5])
plt.ylim([0, 10])

plt.scatter(Fy_EO, Py_EO, color='g', s=5)
```

```python
(Pb_EO, Fb_EO) = welch(df_EO['Balilstogram'], fs=50, window='hann', nperseg=256, nfft=256, noverlap=120)
```

```python
plt.figure(figsize=(10, 4))
plt.title('B_EO')
plt.xlabel(r'Fb_EO')
plt.ylabel(r'Pb_EO')
plt.xlim([-0.0002, 0.002])
plt.ylim([0, 15])

plt.scatter(Fb_EO, Pb_EO, color='g', s=5)
```
